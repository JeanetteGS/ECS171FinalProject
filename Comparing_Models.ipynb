```python
import joblib

log_reg_file = "log_reg_best_model.joblib"
log_reg_model = joblib.load(log_reg_file)

svm_file = "svm_best_model.joblib"
svm_model = joblib.load(svm_file)

mlp_file = "mlp_best_model.joblib"
mlp_model = joblib.load(mlp_file)
```

```python
import matplotlib.pyplot as plt

models = ['LogReg', 'SVC', 'MLP']
accuracies = [log_reg_model['accuracy'], svm_model['accuracy'], mlp_model['accuracy']]
plt.bar(models, accuracies, color='purple')
plt.ylabel('Accuracy Score')
plt.title('Model Type vs. Accuracy Score')
plt.ylim(0, 1)
# Annotate bars
for i, score in enumerate(accuracies):
    plt.text(i, score, f'{score:.4f}', ha='center')
plt.show()
```

```python
import matplotlib.pyplot as plt

models = ['LogReg', 'SVC', 'MLP']
# Class 1 refers to pokemon1 winning combat battle
precisions = [log_reg_model['classification_report']['1']['precision'], svm_model['classification_report']['1']['precision'], mlp_model['classification_report']['1']['precision']]
plt.bar(models, precisions, color='pink')
plt.ylabel('Precision')
plt.title('Model Type vs. Precision Score')
plt.ylim(0, 1)
# Annotate bars
for i, score in enumerate(precisions):
    plt.text(i, score, f'{score:.4f}', ha='center')
plt.show()
```

Precision lets us know if the positives we find are actually true.

```python
import matplotlib.pyplot as plt

models = ['LogReg', 'SVC', 'MLP']
recalls = [log_reg_model['classification_report']['1']['recall'], svm_model['classification_report']['1']['recall'], mlp_model['classification_report']['1']['recall']]
plt.bar(models, recalls, color='skyblue')
plt.ylabel('Recall')
plt.title('Model Type vs. Recall Score')
plt.ylim(0, 1)
# Annotate bars
for i, score in enumerate(recalls):
    plt.text(i, score, f'{score:.4f}', ha='center')
plt.show()
```

Recall tells us if we are missing any positives (false negatives).

```python
import matplotlib.pyplot as plt

models = ['LogReg', 'SVC', 'MLP']
f1s = [log_reg_model['classification_report']['1']['f1-score'], svm_model['classification_report']['1']['f1-score'], mlp_model['classification_report']['1']['f1-score']]
plt.bar(models, f1s, color='orange')
plt.ylabel('F1 score')
plt.title('Model Type vs. F1 Score')
plt.ylim(0, 1)
# Annotate bars
for i, score in enumerate(f1s):
    plt.text(i, score, f'{score:.4f}', ha='center')
plt.show()
```

High F1 score indicates few false positives and few false negatives.
